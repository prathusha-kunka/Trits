```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sc
import pandas as pd
import math
df=pd.read_csv("C:/Users/venkat/Desktop/CSE/IDS/Project/dataset.csv")

amount=list(df['Receipt Amount'])
#print(amount)
meanamount=np.mean(amount)
standamount=np.std(amount)
print(meanamount)
zamount=(meanamount-398000000.881)/(standamount)  #assuming mean amunt to be 398000000
print("P-value of amount is",sc.norm.cdf(zamount)-sc.norm.cdf(-1*zamount))

country=list(df['Donor Name'])
#print(set(country))
amountindex=[]
for i in range(0,len(country)):
    if(country[i]=='Brazil' or country[i]=='India' or country[i]=='Japan' or country[i]=='China'):
        amountindex.append(i)
l=sorted(amountindex)
plt.title('Normal plot of Donor Name')
print("\n")
plt.plot(l,sc.norm.pdf(l,meanamount,standamount))
plt.show()
```
